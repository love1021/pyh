#運算式

```python
a=5
b=3
#透過第三變數c,交換a,b的值
c=a
a=b
b=c
print(a)
print(b) 
      
```

```python
#不透過第三變數 ,交換a,b的值
a=5
b=3
a=a+b
b=a-b
a=a-b
print(a)
print(b)
      
```
